```python
list(df1[df1.Body_Type.isnull()]['Model'].unique())
```

```python
null_models = list(df1[df1.Body_Type.isnull()]['Model'].unique())
```

```python
for i in null_models:
    null_model_indexes = df1[(df1.Model == i) & (df1.Body_Type.isnull())].Body_Type.index
    my_mode = df1[(df1.Model == i)].Body_Type.mode()
    df1.Body_Type.iloc[null_model_indexes]= df1.Body_Type.iloc[null_model_indexes].fillna(my_mode[0])
    print(df1[df1.Body_Type.isnull()].value_counts())  
```
